```python
sampled_cars = cars.sample(1000)
```

```python
sampled_cars[sampled_cars['city'].str.contains('CA')].price.mean() # out of 1000 sampeld cars
```

```python
sampled_cars[sampled_cars['city'].str.contains('TX')].price.mean() # out of 1000 sampeld cars
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

    
# sampled_cars[sampled_cars.year.isnull()]
sampled_cars.dropna(inplace=True) # delete all rows with N/A or null values
# sampled_cars.shape # now down to 129 samples
sampled_cars['year'].astype('int64')


X = sampled_cars[['year']]
y = sampled_cars.price

linreg.fit(X, y);

```

```python

y_pred = linreg.predict(X)
sampled_cars['y_pred'] = y_pred
```

```python
plt.plot(sampled_cars.year , sampled_cars.y_pred) # evaluate year car is made in with price 
```

```python
plt.scatter(sampled_cars.year , sampled_cars.price)
plt.plot(sampled_cars.year , sampled_cars.y_pred, color='red')
plt.xlabel('year')
plt.ylabel('price')
```

```python
sampled_cars['odometer'].astype('int64')

X = sampled_cars[['odometer']]
y = sampled_cars.price

linreg.fit(X, y);
```

```python
y_pred = linreg.predict(X)
sampled_cars['y_pred'] = y_pred
```

```python
plt.plot(sampled_cars.odometer , sampled_cars.y_pred) # price go down as odo goes up
```

```python
plt.scatter(sampled_cars.odometer , sampled_cars.price)
plt.plot(sampled_cars.odometer , sampled_cars.y_pred, color='red')
plt.xlabel('odometer')
plt.ylabel('price')
```
